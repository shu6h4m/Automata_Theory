# Let the PDA for accepting a language {a^n b^n c^n | n>=1 , m>=1}

```python
pda = md2mc('''PDA
I : a, a; aa -> I
I  : a, #; a# -> I
I  : b, a; a -> q1
q1  : b, a; a -> q1
q1  : c, a; ''  -> q2
q2  : c, a; ''  -> q2
q2  : '', #; #  -> F
''')
DOpda = dotObj_pda(pda, FuseEdges=True)
DOpda
```

```python
AnimatePDA(pda, FuseEdges=True)
display(HTML('<link rel="stylesheet" href="//stackpath.bootstrapcdn.com/font-awesome/4.7.0/css/font-awesome.min.css"/>'))
```
