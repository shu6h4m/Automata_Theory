# **Regular Expression** : *A regular expression can be described as a sequence of pattern that defines a string.*


*   The language accepted by finite automata can be easily described by simple expressions called Regular Expressions.
*   It is the most effective way to represent any language.



For Example:

In a regular expression, x* means zero or more occurrence of x. It can generate *{e, x, xx, xxx, xxxx, .....}*

In a regular expression, x+ means one or more occurrence of x. It can generate *{x, xx, xxx, xxxx, .....}*
