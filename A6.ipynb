```python
NFA = md2mc('''
NFA

IF : '' -> A
IF : '' -> B

A : a -> C
B : b -> D 

C : '' -> E
D : '' -> E

E : '' -> IF
E : '' -> G

G : b -> H
H : '' -> E


''')
```

```python
dotObj_nfa(NFA)
```

```python
DFA = nfa2dfa(NFA , STATENAME_MAXSIZE= 50  )
dotObj_dfa(DFA , STATENAME_MAXSIZE=60)
```
